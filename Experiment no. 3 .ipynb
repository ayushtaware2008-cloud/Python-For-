```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
# loading datasets
red=pd.read_csv("winequality-red.csv",sep=";")
white=pd.read_csv("winequality-white.csv",sep=";")
red['type']='red'
white['type']='white'
df=pd.concat([red, white], ignore_index=True)
print(df.head())
```

```python
#lable encoding
le = LabelEncoder()
df['type_label'] = le.fit_transform(df['type'])
print(df[['type','type_label']].head())
```

```python
#one hot encoding
onehot=pd.get_dummies(df['type'], prefix='wine')
df_onehot=pd.concat([df, onehot], axis=1)
print(df_onehot[['type','wine_red','wine_white']].head())
```

```python
# ordinal encoding
ordinal=OrdinalEncoder(categories=[['red','white']])
df['type_ordinal']=ordinal.fit_transform(df[['type']])
print(df[['type','type_ordinal']].head())
```

```python
#Target Encoding
target_mean=df.groupby('type')['quality'].mean()
df['type_target']=df['type'].map(target_mean)
print(df[['type','quality','type_target']].head())
```

```python
#Binary Encoding
df['type_num'] = df['type'].map({'red':0,'white':1})
df['type_binary'] = df['type_num'].apply(lambda x: format(x,'b'))
print(df[['type','type_binary']].head())
```

```python
#Frequency Encoding
freq = df['type'].value_counts(normalize=True)
df['type_frequency'] = df['type'].map(freq)
print(df[['type','type_frequency']].head())
```
