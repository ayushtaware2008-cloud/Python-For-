```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/market_data.csv')
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df['Item_Weight'] = df['Item_Weight'].fillna(df['Item_Weight'].median())
df.isnull().sum()

```

```python
median_visibility = df.loc[df['Item_Visibility'] > 0, 'Item_Visibility'].median()
df['Item_Visibility'] = df['Item_Visibility'].replace(0, median_visibility)
df.drop_duplicates(inplace=True)
Q1 = df['Item_Outlet_Sales'].quantile(0.25)
Q3 = df['Item_Outlet_Sales'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['Item_Outlet_Sales'] >= Q1 - 1.5*IQR) &
        (df['Item_Outlet_Sales'] <= Q3 + 1.5*IQR)]
```
